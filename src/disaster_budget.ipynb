```python
df = pd.read_csv("historical_disaster_data.csv")   
df = df[df["Country"] == "India"]
df.head()
```

```python
hazards = ["Flood", "Storm", "Extreme temperature","Earthquake"]

df = df[df["Disaster Type"].apply(lambda x: x in hazards)]
df.isnull().sum()
```

```python
states = [
    "Maharashtra",
    "Tamil Nadu",
    "Odisha",
    "Assam",
    "Uttar Pradesh",
    "Rajasthan"
]
df = df[df["Location"].str.contains("|".join(states), case=False, na=False)]
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.replace(r"[^\w]", "_", regex=True)
      .str.replace("__+", "_", regex=True)
      .str.strip("_")
)
df = df.rename(columns={
    "total_damage_000_us": "total_damage_usd"
})
```

```python
states = [
    "Maharashtra",
    "Tamil Nadu",
    "Odisha",
    "Assam",
    "Uttar Pradesh",
    "Rajasthan"
]

def extract_state(location):
    for state in states:
        if pd.notna(location) and state.lower() in location.lower():
            return state
    return None

df["state"] = df["location"].apply(extract_state)


df = df[df["state"].notna()]

df["state"].unique()

df.head()
```

```python
df.info()
```

```python
numeric_cols = ["total_damage_usd", "total_deaths", "total_affected"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
def safe_group_median_impute(series):
    # If at least one non-null value exists in the group
    if series.notna().sum() > 0:
        return series.fillna(series.median())
    else:
        return series  # leave NaN for national fallback
```

```python
cols_to_impute = ["total_damage_usd", "total_deaths", "total_affected"]

for col in cols_to_impute:
    df[col] = (
        df.groupby(["state", "disaster_type"])[col]
          .transform(safe_group_median_impute)
    )
```

```python
national_medians = df[cols_to_impute].median()

df[cols_to_impute] = df[cols_to_impute].fillna(national_medians)
```

```python
df.isnull().sum()
```

```python
baseline = (
    df.groupby(["state","disaster_type"])
      .agg(
          total_events=("disaster_type", "count"),
          avg_damage=("total_damage_usd", "median"),
          median_damage=("total_damage_usd", "median"),
          avg_deaths=("total_deaths", "median"),
          avg_affected=("total_affected", "mean")
      )
      .reset_index()
)

baseline.head()
```

```python
events_per_year = (
    df.groupby(["state", "disaster_type", "start_year"])
      .size()
      .reset_index(name="events_in_year")
)

events_per_year.head()
```

```python
frequency = (
    events_per_year.groupby(["state", "disaster_type"])
      .agg(avg_events_per_year=("events_in_year", "mean"))
      .reset_index()
)

frequency.head()
```

```python
baseline = baseline.merge(frequency, on=["state", "disaster_type"], how="left")
```

```python
# --- Safe Min-Max Scaling ---
def safe_min_max_scale(series):
    min_val = series.min()
    max_val = series.max()
    
    if max_val == min_val:
        return pd.Series(0, index=series.index)
    
    return (series - min_val) / (max_val - min_val)


# --- Normalize Core Severity Metrics ---
baseline["damage_score"] = safe_min_max_scale(baseline["avg_damage"])
baseline["deaths_score"] = safe_min_max_scale(baseline["avg_deaths"])
baseline["affected_score"] = safe_min_max_scale(baseline["avg_affected"])
baseline["frequency_score"] = safe_min_max_scale(baseline["avg_events_per_year"])


# --- Financial Vulnerability (Remove frequency from here) ---
baseline["financial_vulnerability"] = baseline["damage_score"]

# --- Human Vulnerability ---
baseline["human_vulnerability"] = (
    0.5 * baseline["deaths_score"] +
    0.5 * baseline["affected_score"]
)

# --- Base Vulnerability (Severity Only) ---
baseline["base_vulnerability"] = (
    0.6 * baseline["financial_vulnerability"] +
    0.4 * baseline["human_vulnerability"]
)

# --- Structural Adjustment (Frequency Multiplier) ---
baseline["historical_vulnerability_index"] = (
    baseline["base_vulnerability"] *
    baseline["frequency_score"]
) * 100


# --- Ranking ---
baseline["risk_rank"] = baseline["historical_vulnerability_index"] \
    .rank(method="dense", ascending=False)


# --- Final Sort ---
baseline = baseline.sort_values(
    "historical_vulnerability_index",
    ascending=False
)

baseline.head(10)
```

```python
final_phase1 = baseline[
    [
        "state",
        "disaster_type",
        "total_events",
        "avg_damage",
        "avg_deaths",
        "avg_affected",
        "avg_events_per_year",
        "historical_vulnerability_index",
        "risk_rank"
    ]
]

final_phase1
```
